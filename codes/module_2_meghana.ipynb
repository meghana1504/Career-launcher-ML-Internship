### QUERY 2.9
### SIMPLE MOVING AVERAGES

```python
mind['Date'] = pd.to_datetime(mind['Date'])
```

```python
fig = plt.figure(figsize=(15,10))
plt.grid()
smas = pd.DataFrame()
smas['signal'] = 0.0
smas['day_21_sma'] = mind['Close Price'].rolling(21).mean() #signals_short
smas['day_34_sma'] = mind['Close Price'].rolling(34).mean() #signals_long

plt.plot(mind['Date'],mind['Close Price'],label='mindtree')
plt.plot(mind['Date'],smas['day_21_sma'],label='21 Days SMA') #short
plt.plot(mind['Date'],smas['day_34_sma'],label='34 Days SMA') #long
plt.title("21 Days SMA and 34 Days SMA on Mindtree")
plt.legend()
```

#### TRADE CALLS CONCLUSION :
    When 21 day SMA crosses 34 day SMA, Buy the mindtree share and vice versa.


### QUERY 2.10
### BOLLINGER BANDS
    1) 14 day rolling mean
    2) Upper band (rolling + 2 std devs)
    3) Lower band (rolling - 2 std devs)
    4) Avg daily prices

```python
plt.figure(figsize=(15,10))
plt.grid()

smas['day_14_sma'] = mind['Close Price'].rolling(14).mean()
std_dev = mind['Close Price'].rolling(14).std()
upper = smas['day_14_sma']+std_dev*2
lower = smas['day_14_sma']-std_dev*2

plt.plot(mind['Date'],mind['Close Price'],label='Mindtree')
plt.plot(mind['Date'],smas['day_14_sma'],label='14 Days SMA')
plt.plot(mind['Date'],upper,label='Upper band')
plt.plot(mind['Date'],lower,label='Lower band')
plt.xlabel("Price")
plt.ylabel("Date")
plt.title("Bollinger plot for 14 days (Mindtree)")

plt.legend()
```

```python
w3 = pd.DataFrame()
w3['Mindtree_close_prices'] = mind['Close Price']
w3['Moving_avg_14'] = smas['day_14_sma']
w3['Std_dev_14'] = std_dev
w3['Upper_band'] = upper
w3['Lower_band'] = lower
w3
```
