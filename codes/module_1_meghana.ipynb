#### 1.4 vwap = sum(price*vol)/sum(volume)


```python
jin['Month'] = pd.DatetimeIndex(jin['Date']).month
jin['Year'] = pd.DatetimeIndex(jin['Date']).year
jin['VWAP'] = (jin['Close Price'] * jin['Total Traded Quantity']).cumsum() / jin['Total Traded Quantity'].fillna(0).cumsum()
jin_vwap = jin[['Month','Year','VWAP']]
group = jin_vwap.groupby(['Month','Year'])
group.first()
```

#### 1.5 Avg price and profit/loss perc over last n days

```python
def avg(n):
    a = np.sum(jin['Close Price'][rows-n:rows])/n
    print("Average: ",a)
```

```python
def prof_loss_perc(n):
    porl = jin['Close Price'][rows-1] -  jin['Close Price'][rows-n]
    if porl < 0:
        loss = porl
        loss_perc = (loss/jin['Close Price'][rows-n])*100
        print("Loss percentage is :",loss_perc*(-1))
    if porl > 0:
        prof = porl
        prof_perc = (prof/jin['Close Price'][rows-n])*100
        print("Profit percentage is :",prof_perc)
```

```python
#1w 2w 1m 3m 6m 1yr
```

```python
print("For one week :")
avg(7)
prof_loss_perc(5)
```

#### 1.7

1.7 Add another column 'Trend' whose values are:
ï€Œ 'Slight or No change' for 'Day_Perc_Change' in between -0.5 and 0.5
ï€Œ 'Slight positive' for 'Day_Perc_Change' in between 0.5 and 1
ï€Œ 'Slight negative' for 'Day_Perc_Change' in between -0.5 and -1
ï€Œ 'Positive' for 'Day_Perc_Change' in between 1 and 3
ï€Œ 'Negative' for 'Day_Perc_Change' in between -1 and -3
ï€Œ 'Among top gainers' for 'Day_Perc_Change' in between 3 and 7

ï€Œ 'Among top losers' for 'Day_Perc_Change' in between -3 and -7
ï€Œ 'Bull run' for 'Day_Perc_Change' >7
ï€Œ 'Bear drop' for 'Day_Perc_Change' <-7

```python
def choices(n):
    if -0.5<=n<=0.5:
        return 'Slight or No change'
    if  0.5<=n<=1:
        return 'Slight positive'
    if -1<=n<=-0.5:
        return 'Slight negative'
    if  1<=n<=3:
        return 'Positive'
    if  -3<=n<=-1:
        return 'Negative'
    if  3<=n<=7:
        return 'Among top gainers'
    if  -7<=n<=-3:
        return 'Among top losers'
    if  n>7:
        return 'Bull run'
    if  n<7:
        return 'Bear run'

jin['Trend'] = jin.Day_Perc_Change.apply(choices)
jin['Trend']
```
